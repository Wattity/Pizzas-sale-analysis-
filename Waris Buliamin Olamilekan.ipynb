## Loading data_set

```python
#importing the data_dictionary dat_set  and printing it
# The dataset is already in the same directory with the notebook so i imported it dirctly
# The data_dictionary is the describe each of the dataset 
data_dictionary=pd.read_csv("data_dictionary.csv")
data_dictionary
```

```python
#importing the order_detail data set and viewing the first five row
# The dataset is already in the same directory with the notebook so i imported it dirctly
order_details= pd.read_csv("order_details.csv")
order_details.head()
```

```python
#importing the order dat_set  and printing the first five row to know what its contain(rows and columns)
# The dataset is already in the same directory with the notebook so i imported it dirctly
orders=pd.read_csv("orders.csv")
orders.head()
```

```python
# merging the orders and order_detail together on order_id column and printing the first five row
merge1=pd.merge(order_details,orders, how="inner", on="order_id")
merge1.head()
```

```python
#importing the order dat_set  and printing the first five row to know what its contain(rows and columns)
# The dataset is already in the same directory with the notebook so i imported it dirctly
pizza=pd.read_csv("pizzas.csv")
pizza.head()
```

```python
# merging the pizzas with merge1(order_details,orders) on pizza_id column and printing the first five row
merge2=pd.merge(merge1,pizza, how="inner", on="pizza_id")
merge2.head()
```

```python
#importing the order dat_set  and printing the first five row to know what its contain(rows and columns)
# The dataset is already in the same directory with the notebook so i imported it dirctly
# using the encoding="latin1" keyword argument because the dataset contain invaid charater coveting it to valid charecter
pizza_type=pd.read_csv("pizza_types.csv",encoding="latin1")
pizza_type.head()
```

```python
# merging the pizzas with merge2 on pizza_type_id column and printing the first five row

data_set=pd.merge(pizza_type,merge2, how="inner", on="pizza_type_id") 
data_set.head()
```

```python
# the date column dtye is object
# converting it to datetime and checking the data_set info to confirm the effect
data_set["date"]=pd.to_datetime(data_set["date"])
data_set.info()
```

```python
#Extracting the months from the date column dataset  and printing the first five row to confirm
data_set["months"]=data_set["date"].dt.month_name()
data_set.head()
```

```python
# extracting the day from the date column dataset  and printing the first five row to confirm
data_set["day"]=data_set["date"].dt.day_name()
data_set.head()
```

```python
#Extracting the day from the date column dataset  and printing the first five row to confirm
data_set["year"]=data_set["date"].dt.year
data_set.head()
```

```python
# Droping the date column since it is already separeted
# # axis=1 argument is use because column is to be drop
# inplace=True to modify our original data_set
data_set.drop("date",axis=1,inplace=True)
data_set.head()
```

```python
# droping the pizza_id 
# There is already size and pizza_type_id column and the pizza_id column is the combination of both and printing the first five row
# axis=1 argument is use because column is to be drop
data_set.drop("pizza_id",axis=1,inplace=True)
data_set.head()
```

## Explortory Data Analysis

### I will be asking the data the following questions

```python
# find the total revenue 
# printing the result
total_revenue=data_set["price"].sum()
print(f"The total revenue: ${total_revenue}")
```

```python
# find the total quantity sold
# printing the result
total_quantity=data_set["quantity"].sum()
print(f"The total quantity sold: {total_quantity}")
```

```python
# find the number of each order and sum it
# printing the result
# NB:the order_id column is in integer finding the sum will not give the total order but the sum of the number in the column
total_order=data_set["order_id"].value_counts().sum()
print(f"The total_order: {total_order}") 
```

```python
# find the total pizzas tyoe sell
# printing the result
# NB: some pizzas type may have duplicate 
# using nunique methold to remve duplcate and give the result in number rather than the list of the pizzas
total_pizzas_type=data_set["pizza_type_id"].nunique()
print(f"The number of pizzas type sell: {total_pizzas_type}") 
```

```python
# find the average price of all pizzas
# printing the result in 2 decimal place
# using $ because the price is input n USD
average_price=data_set["price"].mean()
print(f"The average_price of all pizzas : ${average_price:.2f}") 
```

```python
# find the peek hour of sales
# printing the result
sales_by_hours=data_set.groupby("time")["price"].sum()
print(f" The peek hour of sale: {sales_by_hours.idxmax()}")
```

```python
# find the weekly total sale
# group the day with the price and sum all the price on each day
weekly_total_sales=(data_set.groupby("day")["price"].sum())
print(f"\tThe total sale made on each day of the week:{weekly_total_sales}")
print(f"\t The day of the week when most sale are made: {weekly_total_sales.idxmax()}")
```

```python
top_best_selling_pizzas=data_set.groupby("name")["price"].sum()
#top_best_selling_pizzas.sort_values(ascending=False).head(5)
print(f"\tThe Top Five Selling Pizzas \n{top_best_selling_pizzas.sort_values(ascending=False).head(5)}")
```

```python
# group the day with the price and sum all the price on each day
montly_sales=data_set.groupby("months",as_index=False)["price"].sum()
# converting the months column from object to daytime 
montly_sales["months"]=pd.to_datetime(montly_sales["months"],format="%B").dt.month_name()
#Converting the months column to daytime again but in numeric to use it in arranging the year with nomal calendar
montly_sales["month_num"]=pd.to_datetime(montly_sales["months"],format="%B").dt.month
# sorting the the dataframe with the month and droping it since waht will need it for is sorting
montly_sales=montly_sales.sort_values("month_num").drop(columns="month_num")
# displaying the notice trend from the dataset
print(" The sale is fluctuating,i.e the sale is going up and down without any clear or obvious pattern link to it, below the the break down of the monthly sale")
#set the months as index 
montly_sales.set_index("months")
```

```python
# months with highest sale 
data_set["months"].value_counts()[:1]
```

```python
# group the name of the pizzas with the price and sum all the price on for each name
lowest_sales=data_set.groupby("name")["price"].sum()
#display both the min index and the value
print(f"pizza with the lowest sale is {lowest_sales.idxmin()} with {lowest_sales.min()}")
```

```python
# group the category of the pizzas with the price and sum all the price on for each category
profitable_category=pd.DataFrame(data_set.groupby("category")["price"].sum())
# sorting the category b price from the higest to the lowest
profitable_category.sort_values(by=["price"],inplace=True,ascending=False)
print("\t   The Most Profitable Category")
profitable_category
```
